```python
cat_list = []

for link in links:
    cat = link[40:]
    cat = cat.replace("&sub=", "_")
    cat = cat.replace("%20", " ")
    cat_list.append(cat)
    
cat_list
```

```python
cat_links = dict(zip(cat_list, links))
for cat in cat_links:
    print(cat_links[cat])
```

```python
%%time

title_dict = {}
title_count = 0

for cat in cat_links:
    link = cat_links[cat]
    
    options = webdriver.ChromeOptions()
    headless = True
    
    if headless:       
        options.add_argument("headless")

    driver = webdriver.Chrome('../chromedriver/chromedriver', options=options)
    driver.get(link)
    
    for _ in range(100):
        time.sleep(0.1)
        driver.execute_script("window.scrollTo(9999999999999, 9999999999999);")
        
    title_list = []

    elements = driver.find_elements_by_xpath("//div[@id>100000]")
    for element in elements:
        item_id = str(element.get_attribute("id"))
        title = driver.find_element_by_xpath\
                ('//*[@id="{}"]/div[2]/a'.format(item_id)).text
        title_list.append(title)
        title_count += 1
        
    title_dict[cat] = title_list
    driver.quit()
    
print("총 {}개 타이틀".format(title_count))
print("끄읏")
```

### Hyperparameter tuning
- GridSearchCV

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, make_scorer
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
```

```python
X = df.tokens
y = df.category
```

```python
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [{'vect__ngram_range': [(1, 1), (1, 2), (2, 2)], 
               'clf__C': param_range}]

f1 = make_scorer(f1_score , average='weighted')

pipe_svc = Pipeline([('vect', CountVectorizer()),
                     ('clf', LogisticRegression(class_weight='balanced', random_state=42))])

gs1 = GridSearchCV(estimator=pipe_svc, param_grid=param_grid,
                   scoring=f1, cv=10, n_jobs=6)

%time gs1 = gs1.fit(X, y)
```

```python
import numpy as np

def classes_proba(text, model):
    classes_proba = []
    text = ' '.join(tokenizer.morphs(text))
    prob = model.predict_proba([text])
    prob = np.round(prob, 3).flatten()
    classes = model.classes_
    for cat, prob in zip(classes, prob):
        d = dict()
        d['category'] = cat
        d['value'] = prob
        classes_proba.append(d)
    return classes_proba
```
